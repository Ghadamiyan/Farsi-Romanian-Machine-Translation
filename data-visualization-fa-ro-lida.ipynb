```python
import nltk
import pandas as pd
import numpy as np
from nltk.tokenize import word_tokenize
import string
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
```

```python
train_df = pd.read_csv('../input/ro-fa-clean-translate/train_df.csv')
test_df = pd.read_csv('../input/ro-fa-clean-translate/test_df.csv')
train_df['ro'] = train_df.ro.astype(str).apply(lambda x: x.replace('</i>', ''))
test_df['ro'] = test_df.ro.astype(str).apply(lambda x: x.replace('</i>', ''))
train_df['fa'] = train_df.fa.astype(str).apply(lambda x: x.replace('</i>', ''))
test_df['fa'] = test_df.fa.astype(str).apply(lambda x: x.replace('</i>', ''))
```

```python
train_df['source'].value_counts().plot.bar()
```

```python
train_df['source'].value_counts().plot.pie(subplots=True)
```

```python
v= ' '
for element in train_df['ro']:
    v += str(element)
```

```python
v1 = v.lower()
```

```python
table = str.maketrans(dict.fromkeys(string.punctuation))                   
v2 = (v1.translate(table))
```

```python
words_ = nltk.word_tokenize(v2)
words__=[]
for wr in words_:
    if len(wr) > 3:
        words__.append(wr)
```

```python
from nltk.corpus import stopwords

stopwords = set(stopwords.words('english'))
words = [x for x in words__ if not x in stopwords]  
```

```python
from yellowbrick.text import FreqDistVisualizer

cvect = CountVectorizer(ngram_range=(1, 1), lowercase='true') 
tfidf_transformer = TfidfTransformer(norm= 'l2', use_idf= False)

data1_ = cvect.fit_transform(words)
data_ = tfidf_transformer.fit_transform(data1_)
features = cvect.get_feature_names()

visualizer = FreqDistVisualizer(features=features)
visualizer.fit(data_)
visualizer.poof()
```

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```
